```python
from PIL import Image

from keras.models import load_model
import numpy as np
from numpy import asarray
from numpy import expand_dims

import pickle
import cv2
```

```python
#HaarCascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
HaarCascade = cv2.CascadeClassifier(cv2.samples.findFile(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'))
#MyFaceNet = load_model('facenet_keras.h5')
MyFaceNet = FaceNet()
```

```python
myfile = open("data.pkl", "rb")
database = pickle.load(myfile)
myfile.close()
```

```python
cap = cv2.VideoCapture(0)

while(1):
    _, gbr1 = cap.read()

    wajah = HaarCascade.detectMultiScale(gbr1,1.1,4)

    if len(wajah)>0:
        x1, y1, width, height = wajah[0]
    else:
        x1, y1, width, height = 1, 1, 10, 10

    x1, y1 = abs(x1), abs(y1)
    x2, y2 = x1 + width, y1 + height


    gbr = cv2.cvtColor(gbr1, cv2.COLOR_BGR2RGB)
    gbr = Image.fromarray(gbr)                  # konversi dari OpenCV ke PIL
    gbr_array = asarray(gbr)

    face = gbr_array[y1:y2, x1:x2]

    face = Image.fromarray(face)
    face = face.resize((160,160))
    face = asarray(face)

    #face = face.astype('float32')
    #mean, std = face.mean(), face.std()
    #face = (face - mean) / std

    face = expand_dims(face, axis=0)
    #signature = MyFaceNet.predict(face)
    signature = MyFaceNet.embeddings(face)

    min_dist=100
    identity=' '
    for key, value in database.items() :
        dist = np.linalg.norm(value-signature)
        if dist < min_dist:
            min_dist = dist
            identity = key

    cv2.putText(gbr1,identity, (100,100),cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2, cv2.LINE_AA)
    cv2.rectangle(gbr1,(x1,y1),(x2,y2), (0,255,0), 2)

    cv2.imshow('res',gbr1)

    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()
cap.release()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load your dataset (assuming 'data.pkl' contains labeled face embeddings)
with open("data.pkl", "rb") as myfile:
    database = pickle.load(myfile)

# Prepare your dataset (keys as labels and values as embeddings)
labels = list(database.keys())
embeddings = list(database.values())

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(embeddings, labels, test_size=0.2, random_state=42)

# Convert lists to numpy arrays
X_train = np.array(X_train)
X_test = np.array(X_test)

# Create a function to get the identity from embeddings
def get_identity(embedding, database):
    min_dist = 100
    identity = ' '
    for key, value in database.items():
        dist = np.linalg.norm(value - embedding)
        if dist < min_dist:
            min_dist = dist
            identity = key
    return identity

# Predict identities for the test set
y_pred = [get_identity(embedding, database) for embedding in X_test]

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
